```python
import batman
import numpy as np
import matplotlib.pyplot as plt

# Exoplanet MASCARA-3 Ab 
params1 = batman.TransitParams()       
params1.t0 = 2457146.6                 # time of inferior conjunction
params1.per = 5.5514926                # orbital period
params1.rp = 0.084467                   # planet radius (in units of stellar radii)
params1.a = 9.90312                    # semi-major axis (in units of stellar radii)
params1.inc = 89.16                    # orbital inclination (in degrees)
params1.ecc = 0.085                    # eccentricity
params1.w = 41.                         # longitude of periastron (in degrees)
params1.limb_dark = "quadratic"        # limb darkening model
params1.u = [0.215, 0.29]              # limb darkening coefficients [u1, u2]

# Second exoplanet (with half the radius of the first planet)
params2 = batman.TransitParams()       
params2.t0 = 2457146.6                 # same time of inferior conjunction
params2.per = 5.5514926                # same orbital period
params2.rp = 0.084467 / 2              # planet radius (in units of stellar radii)
params2.a = 9.90312                    # semi-major axis (in units of stellar radii)
params2.inc = 89.16                    # orbital inclination is the same
params2.ecc = 0.085                    # eccentricity is the same
params2.w = 41.                         # longitude of periastron is the same
params2.limb_dark = "quadratic"        # limb darkening model
params2.u = [0.215, 0.29]              # limb darkening coefficients [u1, u2]

# Third exoplanet (with double the radius of the first planet)
params3 = batman.TransitParams()       
params3.t0 = 2457146.6                 # same time of inferior conjunction
params3.per = 5.5514926                # same orbital period
params3.rp = 0.084467 * 2              # planet radius (in units of stellar radii)
params3.a = 9.90312                    # semi-major axis (in units of stellar radii)
params3.inc = 89.16                    # orbital inclination is the same
params3.ecc = 0.085                    # eccentricity is the same
params3.w = 41.                         # longitude of periastron is the same
params3.limb_dark = "quadratic"        # limb darkening model
params3.u = [0.215, 0.29]              # limb darkening coefficients [u1, u2]

# Generate time array
t = np.linspace(0, 1, 500)            #times at which to calculate light curve

# Create light curves for all three planets
m1 = batman.TransitModel(params1, t)    # transit model for first planet
flux1 = m1.light_curve(params1)          # calculate light curve for first planet

m2 = batman.TransitModel(params2, t)    # transit model for second planet
flux2 = m2.light_curve(params2)          # calculate light curve for second planet

m3 = batman.TransitModel(params3, t)    # transit model for third planet
flux3 = m3.light_curve(params3)          # calculate light curve for third planet

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(t, flux1, label='MASCARA-3 Ab', color='cyan')  # First planet (light blue)
plt.plot(t, flux2, label='Half Radius', color='magenta')    # Second planet (fuschia)
plt.plot(t, flux3, label='Double Radius', color='lime') # Third planet (lime green)

plt.xlabel('Time (days)')
plt.ylabel('Relative Flux')
plt.title('Transit Light Curves of Three Planets')
plt.legend()
plt.grid(True)
```
