```python
import numpy as np
import matplotlib.pyplot as plt

filepath = "C:\\Users\\Student\\Documents\\comp_astro_24\\daneel_forward_spectrum.dat"

# Set step size
step = 1000

# Use a generator to read every nth line
wl, depth = [], []

with open(filepath, 'r') as f:
    for i, line in enumerate(f):
        if i % step == 0:
            parts = line.split()
            wl.append(float(parts[0]))
            depth.append(float(parts[1]))

wl = np.array(wl)
depth = np.array(depth)

plt.plot(wl, depth)
plt.xlabel("Wavelength (Âµm)")
plt.ylabel("(Rp/Rs)^2")
plt.show()
```
